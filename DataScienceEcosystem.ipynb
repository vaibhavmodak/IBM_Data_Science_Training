Some of the popular languages that Data Scientists use are:

1. **Python** - A versatile programming language with extensive libraries for data analysis, machine learning, and visualization
2. **R** - A statistical programming language specifically designed for data analysis and statistical computing
3. **SQL** - Essential for database querying and data manipulation in relational databases
4. **Java** - Used for large-scale data processing and enterprise-level applications
5. **Scala** - Popular for big data processing, especially with Apache Spark
6. **Julia** - A high-performance language designed for numerical and scientific computing

Some of the commonly used libraries used by Data Scientists include:

1. **Pandas** - Data manipulation and analysis library for Python, providing data structures like DataFrames
2. **NumPy** - Fundamental package for scientific computing with Python, providing support for arrays and mathematical functions
3. **Matplotlib** - Comprehensive library for creating static, animated, and interactive visualizations in Python
4. **Scikit-learn** - Machine learning library for Python, featuring various classification, regression and clustering algorithms
5. **TensorFlow** - Open-source platform for machine learning and deep learning applications
6. **Seaborn** - Statistical data visualization library based on matplotlib, providing high-level interface for drawing attractive statistical graphics
7. **Plotly** - Interactive graphing library for Python, R, and JavaScript
8. **Keras** - High-level neural networks API for building and training deep learning models
